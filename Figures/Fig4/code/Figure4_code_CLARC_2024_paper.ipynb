```python
funct_labels = {
    'S': 'Unknown function (S)', 'NA': 'No annotation', 'NH': 'No hits',
    'L': 'Replication/Repair (L)', 'P': 'Inorganic ion transport (P)',
    'G': 'Carb metab/transport (G)', 'K': 'Transcription (K)', 'M': 'Cell wall/membrane biogenesis (M)',
    'V': 'Defense (V)', 'E': 'AA metab/transport (E)', 'mixed': 'Mixed',
    'F': 'Nt metab/transport (F)', 'J': 'Translation (J)', 'C': 'Energy production (C)',
    'H': 'Coenzyme metab (H)', 'O': 'Post-trans modif/chaperones (O)',
    'U': 'Intracellular traffic/secretion (U)', 'D': 'Cell cycle/mitosis (D)', 'T': 'Signal transduction (T)',
    'I': 'Lipid transport and metabolism (I)', 'Q': 'Secondary metabolites biosynthesis (Q)',
    'A': 'RNA processing and modification (A)', 'N': 'Cell motility (N)'
}

def load_and_count_cogs(csv_file):
    df = pd.read_csv(csv_file)
    categories = [col.split('_')[-1] for col in df.columns[1:]]
    cog_count = {category: df[col].notna().sum() for col, category in zip(df.columns[1:], categories)}
    return pd.DataFrame(cog_count.items(), columns=['Category', 'COG Count'])

cogs_acc_roary = load_and_count_cogs('~/allcarriage_pangenome_i95/eggnog_annotation/pre-CLARC_acc_genes/functional_accCOG_namelists_roary_i95.csv')
cogs_core_roary = load_and_count_cogs('~/allcarriage_pangenome_i95/eggnog_annotation/pre-CLARC_core_genes/functional_coreCOG_namelists_roary_i95.csv')
cogs_acc_clarc = load_and_count_cogs('~/allcarriage_pangenome_i95/eggnog_annotation/post_CLARC_acc_genes/functional_accCOG_namelists_roaryclarc_i95.csv')
cogs_core_clarc = load_and_count_cogs('~/allcarriage_pangenome_i95/eggnog_annotation/post-CLARC_core_genes/functional_coreCOG_namelists_roaryclarc_i95.csv')

merged_df = cogs_acc_roary.merge(cogs_core_roary, on='Category', suffixes=('_acc_roary', '_core_roary'))
merged_df = merged_df.merge(cogs_acc_clarc, on='Category')
merged_df = merged_df.merge(cogs_core_clarc, on='Category', suffixes=('_acc_clarc', '_core_clarc'))

merged_df['Full Label'] = merged_df['Category'].map(funct_labels).fillna(merged_df['Category'])
merged_df = merged_df.sort_values(by='COG Count_acc_roary', ascending=False)

fig, ax = plt.subplots(figsize=(16, 10))

indices = np.arange(len(merged_df)) * 2
bar_width = 0.7

ax.bar(indices, merged_df['COG Count_acc_roary'], bar_width, color='#A80306', alpha=0.6)
ax.bar(indices + bar_width, merged_df['COG Count_core_roary'], bar_width, color='#676060', alpha=0.6)
ax.bar(indices, merged_df['COG Count_acc_clarc'], bar_width, edgecolor='#2e0202', fill=False, hatch='//', linewidth=1)
ax.bar(indices + bar_width, merged_df['COG Count_core_clarc'], bar_width, edgecolor='#676060', fill=False, hatch='//', linewidth=1)

ax.set_xlabel('COG functional category', fontsize=16)
ax.set_ylabel('COG Count', fontsize=16)
plt.yticks(fontsize=14) 
ax.set_xticks(indices + bar_width / 2)
ax.set_xticklabels(merged_df['Full Label'], rotation=60, ha='right', fontsize=14)

color_handles = [
    Patch(color='#A80306', label='Accessory genes'),
    Patch(color='#676060', label='Core genes')
]

line_handles = [
    Line2D([0], [0], color='black', lw=3, label='Pre-CLARC', alpha=0.6),
    Line2D([0], [0], color='black', lw=2, linestyle='--', label='Post-CLARC', fillstyle='none', marker=None)
]

fig.subplots_adjust(bottom=0.25)

legend1 = ax.legend(handles=color_handles, loc='upper center', bbox_to_anchor=(0.85, 0.95), ncol=2)

legend2 = ax.legend(handles=line_handles, loc='upper center', bbox_to_anchor=(0.85, 0.88), ncol=2)

ax.add_artist(legend1)

plt.tight_layout()

plt.show()
```
